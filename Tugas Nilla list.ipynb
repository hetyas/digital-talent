soal diberika sekumpulan nilai random antara 0 sampai dengan 100 sebanyak 250 nilai
1. tampilkan angka ganjil dari ke 50 nilai tersebut
2. tampilkan berapa banyak yang mendapatkan nilai lebih dari 70
3. tampilkan standar deviase dari sekumpulan nilai tersebut
4. tampilkan nilai maksimun dan minimum
5. tampilkan modus dari sekumpulan nilai tersebut

```python
import random 
#generate random
nilaiTugas =[]
for i in range(250):
    nilaiTugas.append(random.randrange(1,100,1))
    
print(nilaiTugas)
```

```python
#ganjil
for n in nilaiTugas:
    if(n%2==1):
        print (n,"ganjil")
```

```python
#berapa nilai yg lebih dari 70
counter = 0
for x in nilaiTugas:
    if x>=70:
        counter +=1
        print (x)
        print ("jumlah >=70:",counter)
```

```python
#deviasi
jumlah = 0
for i in range (len(nilaiTugas)):
    jumlah += nilaiTugas [i]
    
rata = jumlah/(len(nilaiTugas))
print ("nilai rata-rata:",rata)

sigma = 0
for i in range(len(nilaiTugas)):
    hitung = (nilaiTugas[i]-rata)**2
    sigma += hitung

pembagianN = sigma/(len(nilaiTugas)-1)
SD = pembagianN**0.5
print ("nilai deviasi:",SD)
```

```python
#nilai maks dan min
#nilai maksimum
maks = 0
for ma in nilaiTugas:
    if ma > maks:
        maks=ma
print ("nilai maksimum:",maks)

#nilai minimum
minimum = 0
for mi in nilaiTugas:
    if mi < minimum:
        minimum=mi
print ("nilai minimum:",minimum)
```

```python
#modus
nilai_modus=nilaiTugas [0]
jumlah=nilaiTugas.count(nilai_modus)
for n in nilaiTugas:
    jumlah_n=nilaiTugas.count (n)
    if jumlah_n>jumlah:
        nilai_modus=n
        jumlah=jumlah_n
print (nilai_modus)
```
