```python
import pandas as pd
```

```python
data_melekhuruf = pd.read_csv('processed-angka-melek-huruf-usia-15-24-2005-2009.csv',delimiter=',',encoding="ISO-8859-1")
```

```python
data_melekhuruf
```

Dataset ini berisi mengenai angka melek huruf penduduk berusia 15 - 24 tahun per kabupaten kota. 

```python
data_melekhuruf.count()
```

```python
s = data_melekhuruf.angka_melek_huruf.sum()
mi = data_melekhuruf.angka_melek_huruf.min()
ma = data_melekhuruf.angka_melek_huruf.max()

print ("sum",s)
print ("===")
print ("mins",mi)
print ("===")
print ("maks",ma)
```

```python
me = data_melekhuruf.angka_melek_huruf.mean()
md = data_melekhuruf.angka_melek_huruf.median()
mo = data_melekhuruf.angka_melek_huruf.mode()

print("mean",me)
print ("===")
print("median",md)
print ("===")
print ("modus",mo)
```

```python
#grouping 
data_melekhuruf.groupby('tahun').mean()
```

```python
data_melekhuruf.groupby('nama_provinsi').mean()
```

```python
ax = data_melekhuruf.groupby('tahun').mean()['angka_melek_huruf']
ax = ax.plot(kind='bar',title ="rata-rata berdasarkan tahun", legend=True,fontsize=12)
ax.set_xlabel("tahun",fontsize=12)
ax.set_ylabel("angka melek huruf",fontsize=12)
```

```python
ax = data_melekhuruf.groupby('nama_provinsi').mean()['angka_melek_huruf']
ax = ax.plot(kind='bar',title ="rata-rata berdasarkan provinsi", legend=True,fontsize=12)
ax.set_xlabel("tahun",fontsize=12)
ax.set_ylabel("nama provinsi",fontsize=12)
```

## Narasi analisis
#### visualisasi pertama berdasarkan tahun
bila berdasarkan tahun dapat terlihat peningkatan angka melek huruf di indonesia. setiap tahunnya mengalami peningkatan. 
#### visualisasi kedua berdasarkan provinsi
bila dilihat berdasarkan provinsi maka sangat terlihat bagaimana tingkat melek huruf disetiap provinsinya. terlihat bahwa angka melek huruf tinggi pada daerah seperti provinsi DKI jakarta, jawa tengah. kalimantan dan riau juga baik dalam angka melek huruf. lalu terlihat paling rendah angka melek huruf terjadi di provinsi papua disusul dengan sulawesi barat. dari hasil tersebut makan dapat dilakukan peningatan pendidikan untuk buta huruf di daerah tersebut.
